## Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
- Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ratings.csv')
```

```python
ratings.head()
```

```python
# группируем по юзерам и аггрегируем данные в других колонках: 
# считаем количество рейтингов, выделяем самый ранний и поздний таймстэмп

grouped_by_userId = ratings.groupby('userId').agg({'rating':'count', 'timestamp':['min', 'max']}).reset_index() 
grouped_by_userId
```

```python
# избавляемся от мультииндекса в названиях колонок

grouped_by_userId.columns = ['user_id', 'count_rating', 'timestamp_min', 'timestamp_max']
grouped_by_userId
```

```python
# фультруем пользователей по количеству рейтингов >= 100

users_filtered = grouped_by_userId[grouped_by_userId['count_rating'] >=100]
users_filtered
```

```python
users_filtered['lifetime'] = users_filtered['timestamp_max'] - users_filtered['timestamp_min'] 
users_filtered
```

```python
users_filtered['lifetime_days'] = round(users_filtered['lifetime']/86400,2)
users_filtered
```

```python
mean_lifetime = round(users_filtered['lifetime_days'].mean(),2)
mean_lifetime
```

```python
# с переводом timestamp в datetime
```

```python
ratings['datetime'] = ratings['timestamp'].apply(datetime.fromtimestamp)
ratings
```

```python
grouped_by_userId = ratings.groupby('userId').agg({'rating':'count', 'datetime':['min', 'max']}).reset_index() 
grouped_by_userId.columns = ['user_id', 'count_rating', 'datetime_min', 'datetime_max']
users_filterd = grouped_by_userId[grouped_by_userId['count_rating'] >=100]
users_filterd
```

```python
users_filterd['lifetime'] = users_filterd['datetime_max'] - users_filterd['datetime_min'] 
users_filterd
```

```python
dt_lifetime = users_filterd['lifetime'].mean()
dt_lifetime
```

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента


Обратите внимание, что в процессе объединения таблиц данные не должны теряться.


Дана статистика услуг перевозок клиентов компании по типам:

- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
joined_w_address = (client_base.merge(rzd, how='outer')
                    .merge(auto, how='outer')
                    .merge(air, how='outer')
                    .fillna(0)
                   )

joined_w_address
```

```python
joined = joined_w_address.drop('address' ,axis=1)
joined
```
