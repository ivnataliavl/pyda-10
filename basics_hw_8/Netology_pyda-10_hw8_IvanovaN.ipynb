Добавьте в класс Rate параметр diff (со значениями True или False), который в случае значения True в методах курсов валют (eur, usd итд) будет возвращать не курс валюты, а изменение по сравнению в прошлым значением. Считайте, self.diff будет принимать значение True только при возврате значения курса. При отображении всей информации о валюте он не используется.

```python
import requests
```

```python
class Rate:
    """
    
    """
    def __init__(self, format_='value', diff=False):
        self.format = format_
        self.diff = diff
    
    def exchange_rates(self):
        """
        Возвращает ответ сервиса с информацией о валютах в виде:
        
        {
            'AMD': {
                'CharCode': 'AMD',
                'ID': 'R01060',
                'Name': 'Армянских драмов',
                'Nominal': 100,
                'NumCode': '051',
                'Previous': 14.103,
                'Value': 14.0879
                },
            ...
        }
        """
        self.r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.r.json()['Valute']
    
    def make_format(self, currency):
        """
        Возвращает информацию о валюте currency в двух вариантах:
        - полная информация о валюте при self.format = 'full':
        Rate('full').make_format('EUR')
        {
            'CharCode': 'EUR',
            'ID': 'R01239',
            'Name': 'Евро',
            'Nominal': 1,
            'NumCode': '978',
            'Previous': 79.6765,
            'Value': 79.4966
        }
        
        Rate('value').make_format('EUR')
        79.4966
        
        Изменения для Задания 2
        при self.diff = True возвращается разница между текущим и прошлым курсами одной валюты:
        Rate(diff=True).eur()
        -0.2974000000000103
        
        """
        response = self.exchange_rates()
        
        if currency in response:
            if self.format == 'full':
                return response[currency]
            
            if self.format == 'value':
#   Изменения для Задания 2
                if self.diff:
                    return response[currency]['Value']-response[currency]['Previous']
#   Изменения для Задания 2
                else:
                    return response[currency]['Value']
        return 'Error'
    
    def eur(self):
        """Возвращает курс евро на сегодня в формате self.format"""
        return self.make_format('EUR')
    
    def usd(self):
        """Возвращает курс доллара на сегодня в формате self.format"""
        return self.make_format('USD')
    
    def brl(self):
        """Возвращает курс бразильского реала на сегодня в формате self.format"""
        return self.make_format('BRL')

# Задание 1
    def max_value(self, nominal=False):
        """
        Возвращает название вылюты с максимальным курсом обмена из всего списка валют.
        
        По умолчанию, nominal=False, не учитывает номинал валюты. 
        Функция отдает самое большое значение курса обмена среди всех валют
        
        nominal=True, учитывает номинал и сравнивает курс обмена всех валют за 1 единицу валюты
        """
        response = self.exchange_rates()
        if nominal:
            currency_value = {info['Name']:(info['Value']/info['Nominal']) for valute, info in response.items()}
        else:
            currency_value = {info['Name']:info['Value'] for valute, info in response.items()}
        for name, value in currency_value.items():
            if value == max(currency_value.values()):
                print(f'{name}: {value}')
                return name
            
```

```python
class Employee:
    def __init__(self, name, seniority):
        self.name = name
        self.seniority = seniority
        
        self.grade = 1
    
    def grade_up(self):
        """Повышает уровень сотрудника"""
        self.grade += 1
    
    def publish_grade(self):
        """Публикация результатов аккредитации сотрудников"""
        print(self.name, self.grade)
    
    def check_if_it_is_time_for_upgrade(self):
        pass
```

```python
class Developer(Employee):
    def __init__(self, name, seniority):
        super().__init__(name, seniority)
    
    def check_if_it_is_time_for_upgrade(self):
        # для каждой аккредитации увеличиваем счетчик на 1
        # пока считаем, что все разработчики проходят аккредитацию
        self.seniority += 1
        
        # условие повышения сотрудника из презентации
        if self.seniority % 5 == 0:
            self.grade_up()
        
        # публикация результатов
        return self.publish_grade()
```

```python
class Designer(Employee):
    
    def __init__(self, name, seniority, awards):
        super().__init__(name, seniority)
        
        self.awards = awards
    
    def seniority_upgrade(self, passed=True):
        """
        Счетчик пройденных аккредитаций. 
        Если аккредитация пройдена - параметр passed=True. 
        По умолчанию считается, что аккредитация пройдена.
        """
        if passed:
            self.seniority += 1
            print('Employee passed the accreditation!')
        else: 
            print('Employee didn\'t pass')
    
    def new_award(self, win_award=True):
        """
        Счетчик выигранных наград.
        Если награда выиграна - параметр award=True.
        По умолчанию сичтается, что награда получена.
        """
        if win_award:
            self.awards += 1
            print('Employee won award!')
        else:
            print('Employee didn\'t win')
            
    def check_if_it_is_time_for_upgrade(self):
        """
        Проверяет на совпадение условий по повышению грейда.
        Первая проверка - на изменение количества аккредитаций и наград.
        Вторая проверка - на соответствие условию повышения - за каждые 7 баллов
        """
        self.total = self.seniority + (self.awards*2)
        
        passed = input('Has employee passed the accreditation? True/False ')
        self.seniority_upgrade(passed)
        
        win_award = input('Has employee won new award? True/False ')
        self.new_award(win_award)
        
        if self.total != self.seniority + (self.awards*2): #проверка: увеличились ли количество аккредитаций и наград
            self.total = self.seniority + (self.awards*2) 
            if self.total // 7 > self.grade: #проверка на соответствие условию повышения
                self.grade_up()
                print(f'Total points: {self.total}') 
                print(f'Accreditations: {self.seniority}, awards: {self.awards}. Grade up! Current grade {self.grade}')
            else:
                print(f'Total points: {self.total}') 
                print(f'Accreditations: {self.seniority}, awards: {self.awards}. No grade up! Current grade {self.grade}')
            
            
```
