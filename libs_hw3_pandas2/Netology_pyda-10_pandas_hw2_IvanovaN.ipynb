## Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class


```python
film_ratings = pd.read_csv('ratings.csv')
```

```python
film_ratings.head()
```

```python
def classify_by_rating(rating):
    if rating <= 2:
        return 'low'
    elif rating <= 4:
        return 'average'
    return 'high'
```

```python
film_ratings['class'] = film_ratings['rating'].apply(classify_by_rating)
film_ratings.head()
```

## Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.


```python
keywords = pd.read_csv('keywords.csv')
keywords
```

```python
def geo_classification(keyword):
    geo_data = {

        'Центр': ['москва', 'тула', 'ярославль'],

        'Северо-Запад': ['петербург', 'псков', 'мурманск'],

        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    
    for word in keyword.split(' '):
        for region, cities in geo_data.items():
            if word in cities:
                return region 
    return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(geo_classification)
```

```python
keywords[keywords['region'] == 'Центр'].head()
```

```python
keywords[keywords['region'] ==  'Северо-Запад'].head()
```

```python
keywords[keywords['region'] ==  'Дальний Восток'].head()
```

```python
keywords[keywords['region'] ==  'undefined'].head()
```

## Задание 3 (бонусное)
Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

- В переменную years запишите список из всех годов с 1950 по 2010.

- Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

        - для каждой строки пройдите по всем годам списка years
        - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
        - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
- Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

- Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

**Выгрузка документов**

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

**Считаем среднюю оценку по каждому фильму**

```python
grouped_movies = ratings[['movieId', 'rating']].groupby('movieId').mean().reset_index()
grouped_movies
```

**Добавляем названия в датафрейм со средней оценкой по фильмам**

```python
movies_w_ratings = movies.merge(grouped_movies[['movieId', 'rating']], on='movieId')
movies_w_ratings = movies_w_ratings[['movieId', 'title', 'rating']]
movies_w_ratings
```

**Выделяем год из названия**

```python
import re
```

```python
def get_year(title):
    year_exp = r'[?:(](\d{4})(?:-*)(?:\d+)*[?:)]'
    year = re.findall(year_exp, title)
    if year:
        return year[0]
    return '1900'
```

```python
movies_w_ratings['years'] = movies_w_ratings['title'].apply(get_year)
```

```python
movies_w_ratings
```

**Делаем группировку по годам и считаем средний рейтинг фильмов**

```python
avg_rating_by_years = movies_w_ratings[['years', 'rating']].groupby('years').mean().sort_values('rating', ascending=False).reset_index()
```

```python
avg_rating_by_years.head(10)
```

```python
avg_rating_by_years[::-1].head(10)
```

**Вывод:**
Если не брать в расчет факторы влияющие на выставленную оценку, количество выпускаемых фильмов и прочие факторы, то утверждение  “раньше снимали настоящее кино, не то что сейчас” скорее верное чем ложное. 
Среди годов с самым высоким средним рейтингов фильмов только года до 1930, с самым низким - большая часть позже 2000-х

Добавим к каждому фильму классификацию рейтинга (функция из 1 задания) - low, high, average

```python
movies_w_ratings['label'] = movies_w_ratings['rating'].apply(classify_by_rating)
movies_w_ratings
```

Посчитаем сколько фильмов каждого лейбла снимали в разрезе годов

```python
labeled_films = movies_w_ratings[['years', 'rating', 'label']].groupby(['years', 'label']).size().unstack(fill_value=0)

labeled_films
```

Разделим на 2 группы "старые" и "современные"

```python
labeled_films[:'1979']
```

```python
labeled_films['1980':]
```

Посчитаем среднее количество фильмов с низким и высоким рейтингом среди современных и старых фильмов

```python

modern_low_rate_avg = labeled_films['1980':]['low'].mean()
modern_low_rate_avg
```

```python
modern_high_rate_avg = labeled_films['1980':]['high'].mean()
modern_high_rate_avg
```

```python
old_low_rate_avg = labeled_films[:'1979']['low'].mean()
old_low_rate_avg
```

```python
old_high_rate_avg = labeled_films[:'1979']['high'].mean()
old_high_rate_avg
```

Посчитаем соотношение среднего количества фильмов с низким рейтингом к фильмам с высоким рейтингом внутри каждой группы

```python
modern_low_high_rate_ratio = modern_low_rate_avg / modern_high_rate_avg
modern_low_high_rate_ratio
```

```python
old_low_high_rate_ratio = old_low_rate_avg / old_high_rate_avg
old_low_high_rate_ratio
```

```python
modern_low_high_rate_ratio / old_low_high_rate_ratio
```
